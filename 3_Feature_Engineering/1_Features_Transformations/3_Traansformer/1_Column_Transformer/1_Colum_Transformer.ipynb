# Column Transformer

```python
X_train ,X_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('trf1',SimpleImputer(),['fever']),
    ('trf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('trf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
X_train_trf = transformer.fit_transform(X_train)
X_test_trf = transformer.transform(X_test)
```
