```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df = df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=0)
```

```python
X_train
```

# check null

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer()
si_cat = SimpleImputer(strategy='most_frequent')

X_train_age = si.fit_transform(X_train[['Age']])
X_test_age = si.transform(X_test[['Age']])

X_train_embarked = si_cat.fit_transform(X_train[['Embarked']])
X_test_embarked = si_cat.transform(X_test[['Embarked']])
```

```python
pd.DataFrame(X_train_age).isnull().sum()
pd.DataFrame(X_train_embarked).isnull().sum()
```

# num data -> standarization

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X_train_fair = ss.fit_transform(X_train[['Fare']])
X_test_fair = ss.transform(X_test[['Fare']])

X_train_age = ss.fit_transform(X_train_age)
X_test_age = ss.transform(X_test_age)
```

```python
X_train_fair
```

```python
X_train_age
```

# cat-> Encoding (Ordinal or onehot)

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore')

X_train_sex = ohe.fit_transform(X_train[['Sex']])
X_test_sex = ohe.transform(X_test[['Sex']])

X_train_embarked = ohe.fit_transform(X_train_embarked)
X_test_embarked = ohe.transform(X_test_embarked)
```

```python
X_train_sex
```

```python
train_rest = X_train[['Pclass','SibSp','Parch']]
test_rest = X_test[['Pclass','SibSp','Parch']]
train_rest
```

```python
print(train_rest.shape)
print(X_train_sex.shape)
print(X_train_age.shape)
print(X_train_embarked.shape)
print(X_train_fair.shape)
```

```python
X_train_trf = np.concatenate((X_train_age,X_train_sex,X_train_embarked,X_train_fair,train_rest),axis=1)
X_test_trf = np.concatenate((X_test_age,X_test_sex,X_test_embarked,X_test_fair,test_rest),axis=1)
```

```python
X_train_trf
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train_trf,y_train)
y_pred = dtc.predict(X_test_trf)
accuracy_score(y_test,y_pred)
```
