```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer.csv',usecols=['review','education','purchased'])
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# For categorical ordinal data 

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
X_train_encoded = oe.fit_transform(X_train)
X_test_encoded = oe.transform(X_test)
```

```python
X_train_encoded
```

```python
X_train_encoded = pd.DataFrame(X_train_encoded,columns=X_train.columns)
X_train_encoded
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(16,6))
sns.histplot(data=X_train,x='review',ax=axes[0])
sns.histplot(data=X_train_encoded,x='review',ax=axes[1])
```

# Label Encoding for y

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```
