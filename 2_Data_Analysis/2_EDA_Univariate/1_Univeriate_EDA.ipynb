```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
football = pd.read_csv('2021-2022 Football Player Stats.csv',sep=';',encoding='latin-1',usecols=['Player','Nation','Pos','Squad','Comp','Age','Born','MP','Goals','Shots','SoT','Assists','Off','CrdR','CrdY','2CrdY'])
football = football.sample(300)
football
```

### a. Numerical

# 1. Histplot

```python
sns.displot(kind='hist',data=football,x='Age')
```

```python
sns.displot(kind='hist',data=football,x='Shots')
```

# 2. KdePlot

```python
sns.displot(kind='kde',data=football,x='Age')
```

```python
sns.displot(kind='kde',data=football,x='Age',hue='Comp',fill=True)
```

```python
sns.displot(kind='kde',data=football,x='Goals')
sns.rugplot(data=football,x='Goals')
```

# 3. Boxplot 

```python
sns.catplot(kind='box',data=football,x='Age')
```

```python
sns.catplot(kind='box',data=football,x='MP')
```

### b. Categorical

# 1. Count plot

```python
football
```

```python
sns.catplot(kind='count',data=football,x='Comp')
```

```python
sns.catplot(kind='count',data=football,x='Comp',hue='Pos')
```

# 2. Pie Chart

```python
football
```

```python
plt.figure(figsize=(15,15))
football['Pos'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
plt.figure(figsize=(15,15))
football['Nation'].value_counts().plot(kind='pie',autopct='%.2f')
```
