```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=4, n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(X,y)
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.intercept_, reg.coef_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

## Lets apply Gradient Descent assuming slope is constant m = 78.35 and b = 100

```python
y_pred = ((78.35 * X) + 100)
y_pred.shape
```

```python
X
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=100')
plt.legend()
```

##### iteration 1 

```python
m = 78.5
b = 100

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1
b = b - (lr * loss_slope)
b
```

```python
y_pred1 = ((78.35 * X) + b)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=100')
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b))
plt.legend()
```

##### iteration 2

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
lr = 0.1
b = b - (lr * loss_slope)
b
```

```python
y_pred2 = ((78.35 * X) + b)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b=100')
plt.plot(X, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(X, y_pred2, color='#3AAFB9', label='b={}'.format(b))
plt.legend()
```

##### iteration 3

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
lr = 0.1
b = b - (lr * loss_slope)
b
```

```python
y_pred3 = ((78.35 * X) + b)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='b=100')
plt.plot(X, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(X, y_pred2, color='#3AAFB9', label='b={}'.format(b))
plt.plot(X, y_pred3, color='#9C914F', label='b={}'.format(b))
plt.legend()
```

```python
import random

def generate_random_hex_color():
    # Generate a random integer between 0 and 0xFFFFFF (16777215 in decimal)
    random_color = random.randint(0, 0xFFFFFF)

    # Convert the integer to a hexadecimal string and format it to 6 digits
    hex_color = f'#{random_color:06x}'

    return hex_color

# Example usage
random_hex = generate_random_hex_color()
print(random_hex)
```

```python
plt.figure(figsize=(16,12))
b = -100 
m = 78.35
lr = 0.1

epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slope)
    
    y_pred = m * X + b
    plt.plot(X,y_pred,color='{}'.format(generate_random_hex_color()),label='itr{}'.format(i))
plt.scatter(X,y)
plt.legend()
    
```
